```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/amazon_review.csv'
df = pd.read_csv(file_path)

# Convert 'reviewTime' to datetime format
df['reviewTime'] = pd.to_datetime(df['reviewTime'])

# Extract year and month for analysis
df['year'] = df['reviewTime'].dt.year
df['month'] = df['reviewTime'].dt.month

# Plot distribution of ratings
plt.figure(figsize=(8,5))
sns.countplot(x='overall', data=df, palette='viridis')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Plot number of reviews over time
plt.figure(figsize=(12,5))
df['year'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Reviews Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.show()

# Helpful votes vs. total votes
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['total_vote'], y=df['helpful_yes'], alpha=0.5)
plt.title('Helpful Votes vs. Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('Helpful Votes')
plt.show()

# Monthly trend of reviews
df.groupby(['year', 'month']).size().unstack().T.plot(figsize=(12,5), cmap='coolwarm')
plt.title('Monthly Review Trend')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.legend(title='Year', bbox_to_anchor=(1,1))
plt.show()

# Wordcloud for frequent words in reviews (optional, if needed)
from wordcloud import WordCloud
text = " ".join(review for review in df['reviewText'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Frequent Words in Reviews')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/amazon_review.csv'
df = pd.read_csv(file_path)

# Convert 'reviewTime' to datetime format
df['reviewTime'] = pd.to_datetime(df['reviewTime'])

# Extract year, month, and season for analysis
df['year'] = df['reviewTime'].dt.year
df['month'] = df['reviewTime'].dt.month
df['season'] = df['month'] % 12 // 3 + 1  # 1: Winter, 2: Spring, 3: Summer, 4: Fall

# Seasonal analysis
plt.figure(figsize=(8,5))
sns.countplot(x='season', data=df, palette='coolwarm')
plt.title('Reviews by Season')
plt.xlabel('Season (1: Winter, 2: Spring, 3: Summer, 4: Fall)')
plt.ylabel('Number of Reviews')
plt.show()

# Monthly trend of reviews
df.groupby(['year', 'month']).size().unstack().T.plot(figsize=(12,5), cmap='coolwarm')
plt.title('Monthly Review Trend')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.legend(title='Year', bbox_to_anchor=(1,1))
plt.show()

# Purchase behavior analysis
plt.figure(figsize=(8,5))
sns.countplot(x='overall', data=df, palette='viridis')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Purchase trends over time
plt.figure(figsize=(12,5))
df['year'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Reviews Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.show()

# Helpful votes vs. total votes
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['total_vote'], y=df['helpful_yes'], alpha=0.5)
plt.title('Helpful Votes vs. Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('Helpful Votes')
plt.show()

# Wordcloud for frequent words in reviews (optional, if needed)
from wordcloud import WordCloud
text = " ".join(review for review in df['reviewText'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Frequent Words in Reviews')
plt.show()
```
